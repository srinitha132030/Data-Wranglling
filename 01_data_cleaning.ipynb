## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

## LOAD THE DATASET

```python
sales2020 = pd.read_csv("C:/Users/DELL/OneDrive/Desktop/apexplanet2/dataset/raw/AdventureWorks Sales Data 2020.csv")
sales2021 = pd.read_csv("C:/Users/DELL/OneDrive/Desktop/apexplanet2/dataset/raw/AdventureWorks Sales Data 2021.csv")
sales2022 = pd.read_csv("C:/Users/DELL/OneDrive/Desktop/apexplanet2/dataset/raw/AdventureWorks Sales Data 2022.csv")
```

```python
print("Sales 2020 Shape:", sales2020.shape)
print("Sales 2021 Shape:", sales2021.shape)
print("Sales 2022 Shape:", sales2022.shape)
```

## PREVIEW THE DATA

```python
sales2020.head()
sales2021.head()
sales2022.head()
```

##  CHECKING COLUMN CONSISTENCY

```python
print(sales2020.columns.tolist())
print(sales2021.columns.tolist())
print(sales2022.columns.tolist())## TO LIST COVERTS CERTAIN DATASTRUCTURES INTO NORMAL PYTHON
```

## MERGING SALES DATA

```python
sales_all = pd.concat(
    [sales2020, sales2021, sales2022],
    ignore_index=True
)
```

```python
print("Combined Dataset Shape:", sales_all.shape)
sales_all.head()
```

## DATASET INFORMATION

```python
sales_all.info()
```

## MISSING VALUES

```python
missing_values = sales_all.isnull().sum()

missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Values': missing_values.values
})

missing_df[missing_df['Missing Values'] > 0]
```

```python
missing_percentage = (
    sales_all.isnull().sum()
    / len(sales_all)
) * 100

missing_percentage.sort_values(
    ascending=False
)
```

## DUPLICATED RECORDS

```python
duplicates = sales_all.duplicated().sum()

print(
    f"Duplicate Rows: {duplicates}"
)
```

```python
sales_all.drop_duplicates(
    inplace=True
)
```

```python
sales_all.duplicated().sum()
```

## DATA TYPE CONVERSION

```python
sales_all.columns
```

```python
sales_all['OrderDate'] = pd.to_datetime(
    sales_all['OrderDate']
)
```
