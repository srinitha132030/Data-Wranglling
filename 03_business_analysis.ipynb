```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
sales = pd.read_csv("../data/processed/Sales_All.csv")

sales.head()
```

```python
print(sales.columns.tolist())
```

```python
total_orders = len(sales)

print("Total Orders:", total_orders)
```

```python
total_quantity = sales["OrderQuantity"].sum()

print("Total Quantity Sold:", total_quantity)
```

```python
top_products = (
    sales.groupby("ProductKey")
    ["OrderQuantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

```python
top_products.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Products")
plt.xlabel("Product Key")
plt.ylabel("Quantity Sold")

plt.show()
```

```python
territory_sales = (
    sales.groupby("TerritoryKey")
    ["OrderQuantity"]
    .sum()
    .sort_values(ascending=False)
)

print(territory_sales)
```

```python
territory_sales.head(10).plot(
    kind="bar"
)

plt.title("Top Territories")
plt.show()
```
