## HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv("../data/processed/Sales_All.csv")

sales["OrderQuantity"].hist(bins=20)

plt.title("Distribution of Order Quantity")
plt.xlabel("Order Quantity")
plt.ylabel("Frequency")

plt.show()
```

## BAR CHART

```python
import matplotlib.pyplot as plt

top_products = (
    sales["ProductKey"]
    .value_counts()
    .head(10)
)

top_products.plot(kind="bar")

plt.title("Top 10 Products by Orders")
plt.xlabel("Product Key")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

plt.show()
```

## HEAT MAP

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
sales = pd.read_csv("../data/processed/Sales_All.csv")

# Select only numeric columns
numeric_data = sales.select_dtypes(include='number')

# Correlation matrix
corr_matrix = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

## SCATTER PLOT

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    sales["OrderLineItem"],
    sales["OrderQuantity"]
)

plt.title("Order Line Item vs Order Quantity")
plt.xlabel("Order Line Item")
plt.ylabel("Order Quantity")

plt.show()
```
